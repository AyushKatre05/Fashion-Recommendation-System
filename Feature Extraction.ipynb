```python
import cv2
import numpy as np
from numpy.linalg import norm

def extract_feature(img_path, model):
    img = cv2.imread(img_path)
    img = cv2.resize(img, (720, 540))
    img = np.array(img)
    expand_img = np.expand_dims(img, axis=0)
    pre_img = preprocess_input(expand_img)
    result = model.predict(pre_img).flatten()  # Corrected the flatten call
    normalized = result / norm(result)
    return normalized

# Call the function
extract_feature("data/Footwear/Men/Images/images_with_product_ids/1636.jpg", model)
```

```python
def extract_feature(img_path, model):
    try:
        img = cv2.imread(img_path)
        if img is None:
            print(f"Error: Unable to read image '{img_path}'. Skipping...")
            return None
        img = cv2.resize(img, (224, 224))  # Resize to match ResNet50 input size
        img = np.array(img)
        expand_img = np.expand_dims(img, axis=0)
        pre_img = preprocess_input(expand_img)
        result = model.predict(pre_img).flatten()
        normalized = result / norm(result)
        return normalized
    except Exception as e:
        print(f"Error processing '{img_path}': {e}")
        return None
```

```python
import os

# Define data directories for different categories
men_directory = r'C:/Users/ayush/OneDrive/Desktop/Fashion Recommendation System/data/Footwear/Men/Images/images_with_product_ids'
women_directory = r'C:/Users/ayush/OneDrive/Desktop/Fashion Recommendation System/data/Footwear/Women/Images/images_with_product_ids'
boys_directory = r'C:/Users/ayush/OneDrive/Desktop/Fashion Recommendation System/data/Apparel/Boys/Images/images_with_product_ids'
girls_directory = r'C:/Users/ayush/OneDrive/Desktop/Fashion Recommendation System/data/Apparel/Girls/Images/images_with_product_ids'

# List of directories containing images
data_directories = [men_directory, women_directory, boys_directory, girls_directory]

men_filenames = []
women_filenames = []
boys_filenames = []
girls_filenames = []

# Collect paths of all images in the specified directories for each category
for directory in data_directories:
    for root, _, files in os.walk(directory):
        for file in files:
            full_path = os.path.join(root, file)
            if directory == men_directory:
                men_filenames.append(full_path)
            elif directory == women_directory:
                women_filenames.append(full_path)
            elif directory == boys_directory:
                boys_filenames.append(full_path)
            elif directory == girls_directory:
                girls_filenames.append(full_path)
```

```python
from tqdm import tqdm

feature_list = []

# Iterate over all paths and extract features
for file in tqdm(filename):
    try:
        feature_list.append(extract_feature(file, model))
    except Exception as e:
        print(f"Error processing file '{file}': {e}")
```
